```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.stats.api as sms
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from math import ceil
from statsmodels.stats.proportion import proportions_ztest, proportion_confint                 
```

```python
#Загрузка датасета
df = pd.read_csv(r'C:\Users\Herman\Study\Data Science Course(it-by)\homework_2\ab_test_inference\ab_test_ensembles.csv')
df.head()
```

```python
# Проверка распределения групп и вариантов 
pd.crosstab(df['group'], df['variants'])
```

```python
# Data Cleaning
# Позьлователи, которые попали в выборку несколько раз
sessions = df['user_id'].value_counts(ascending=False)
duplicate_users = sessions[sessions > 1].count()
print(f"\n Всего {duplicate_users} пользователей, которые не уникальны.")

#Удаляем дуплицитные значения
users_to_drop = sessions[sessions > 1].index
df = df[~df['user_id'].isin(users_to_drop)]

# Оставляем только корректные соответствия
df = df[((df['group'] == 'control') & (df['variants'] == 'standard')) | 
        ((df['group'] == 'treatment') & (df['variants'] == 'svm_ensemble_v_1_22'))]
print(f"Количество пользователей после очистки: {df.shape[0]}")
pd.crosstab(df['group'], df['variants'])
```

```python
# A/B test
# Расчет Power Analysis с учетом роста конверсии на 2%(с 12 до 14)

effect_size = sms.proportion_effectsize(0.12, 0.14)
required_n = sms.NormalIndPower().solve_power(effect_size, power=0.8, alpha=0.05, ratio=1)
required_n = ceil(required_n)
print(f"\nДля каждой группы необходимо не менее {required_n} наблюдений.")
```

```python
#Формирование ab_df через случайную выборку required_n 
control_sample = df[df['group'] == 'control'].sample(n=required_n, random_state=42)
treatment_sample = df[df['group'] == 'treatment'].sample(n=required_n, random_state=42)

ab_df = pd.concat([control_sample, treatment_sample], axis=0)
ab_df.reset_index(drop=True, inplace=True)

ab_df.group.value_counts(normalize=True)
```

```python
#Описательная статистика
conversion_rates = ab_df.groupby('group')['converted']

std_proportion = lambda x: np.std(x, ddof=0)
se_proportion = lambda x: stats.sem(x, ddof=0)

conversion_rates = conversion_rates.agg(
    ['mean', std_proportion, se_proportion]
).rename(
    columns={'mean': 'CR mean', '<lambda_0>': 'CR std', '<lambda_1>': 'CR se'}
)

conversion_rates
```

```python
#Проверка гипотез (Z test & CI)
from statsmodels.stats.proportion import proportions_ztest, proportion_confint

def get_inference(data, label):
    control_results = data[data['group'] == 'control']['converted']
    treatment_results = data[data['group'] == 'treatment']['converted']

    n_con = control_results.count()
    n_treat = treatment_results.count()
    successes = [control_results.sum(), treatment_results.sum()]
    nobs = [n_con, n_treat]

    z_stat, pval = proportions_ztest(successes, nobs=nobs)
    (lower_con, lower_treat), (upper_con, upper_treat) = proportion_confint(successes, nobs=nobs, alpha=0.05)

    print(f"\n--- {label} ---")
    print(f"Z-statistic: {z_stat:.2f}")
    print(f"p-value: {pval:.3f}")
    print(f"CI 95% for control group:   [{lower_con:.3f}, {upper_con:.3f}]")
    print(f"CI 95% for treatment group: [{lower_treat:.3f}, {upper_treat:.3f}]")

    if pval < 0.05:
        print("Результат: Различия статистически значимы.")
    else:
        print("Результат: Различия статистически НЕ значимы.")
```

```python
#Ответ на вопрос 1 : Заменять старый вариант не стоит
get_inference(df, "Общий результат по всей выборке")
```

```python
#Ответ на вопрос 2: 
#Для registered_user статистически значимых различий нет, можно оставить старый вариант.
#Для new_user различия значимы, следует использовать старую модель.
улучшения конверсии для этого сегмента
for utype in df['user_type'].unique():
    get_inference(df[df['user_type'] == utype], f"СЕГМЕНТ: {utype}")
```
