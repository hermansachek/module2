```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-whitegrid")
sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 100)
```

```python
df = pd.read_csv(r'C:\Users\Herman\Study\Data Science Course(it-by)\homework_2\descriptive_stats\clients.csv')
df.head()

df.columns = df.columns.str.strip()

df.info()
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
print("Unique customers:", df["Customer ID"].nunique())
```

```python
#Missing values 
missing_cnt = df.isna().sum().sort_values(ascending=False)
missing_pct = (df.isna().mean() * 100).sort_values(ascending=False).round(2)

missing_table = pd.DataFrame({"missing_cnt": missing_cnt, "missing_%": missing_pct})
missing_table[missing_table["missing_cnt"] > 0]
```

```python
#Data Cleaning

def parse_price(x):
    x = str(x).replace("$", "").replace(",", "").strip()
    return float(x)

df["Price_num"] = df["Price"].apply(parse_price)
df["Area_sqm"] = df["Area (ft.)"] * 0.092903
df["Price_per_sqm"] = df["Price_num"] / df["Area_sqm"]

# пропуски в категориальных атрибутах заменяем на "Unknown"
df["Gender"] = df["Gender"].fillna("Unknown")
df["State"] = df["State"].fillna("Unknown")
df["Age Interval"] = df["Age Interval"].fillna("Unknown")

#  пропуски в атрибуте "Age at time of purchase" заменяем медаианой
df["Age at time of purchase"] = df["Age at time of purchase"].fillna(df["Age at time of purchase"].median())

df[["Price_num", "Area_sqm", "Price_per_sqm"]].head()
```

```python
#Делаем Correlation heatmap
num_cols = ["Area_sqm", "Price_num", "Age at time of purchase", "Deal satisfaction", "Price_per_sqm"]
corr = df[num_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation heatmap (numeric features)")
plt.show()
```

```python
# EDA - графики

#1 - Age dsitribution
plt.figure(figsize=(8, 5))
sns.histplot(df["Age at time of purchase"], bins=15, kde=True)
plt.title("Age distribution of clients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
Смотрим как распределены покупатели по возрастным группам. Можно понять люди в каком
возрастном диапазоне чаще покупают недвижимость.
```

```python
#2 - Gender distribution

plt.figure(figsize=(6, 4))
sns.countplot(x="Gender", data=df, order=df["Gender"].value_counts().index)
plt.title("Gender distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Смотрим на гендерное разделение покупателей. Это может влиять на способы продвижения недвижимости при
продажах, тональность коммуникации.

```python
#3 - Purchase Purpose

plt.figure(figsize=(7, 4))
sns.countplot(x="Purpose", data=df, order=df["Purpose"].value_counts().index)
plt.title("Purpose of purchase")
plt.xlabel("Purpose")
plt.ylabel("Count")
plt.show()
```

Смотрим на разделение по цели покупки. В зависимости от цели покупки люди будут предпочитать
те или иные варианты недвижимости. Например, преимущества, актуальные для квартиры, 
купленные для инвестиции могут быть не актуальны для домов, купленных для семейного проживания. 

```python
#4 - Lead source 

plt.figure(figsize=(8, 4))
sns.countplot(x="Source", data=df, order=df["Source"].value_counts().index)
plt.title("Lead sources (Source)")
plt.xlabel("Source")
plt.ylabel("Count")
plt.xticks(rotation=25)
plt.show()
```

Ключевой параметр для настройки рекламных каналов.

```python
#5 - Price depending on purpose

plt.figure(figsize=(8, 5))
sns.boxplot(x="Purpose", y="Price_num", data=df)
plt.title("Price distribution by purchase purpose")
plt.xlabel("Purpose")
plt.ylabel("Price ($)")
plt.show()
```

Распределение цен в зависимости от цели покупки. Полезно для понимания рынка. В каком ценовом диапазоне люди более склонны к покупке недвижимости для жизни, а в каком для инвестиций. 

```python
#6 - Purpose depending on age

plt.figure(figsize=(8, 5))
sns.countplot(x="Age Interval", hue="Purpose", data=df)
plt.title("Purchase purpose by age interval")
plt.xlabel("Age interval")
plt.ylabel("Count")
plt.xticks(rotation=25)
plt.show()
```

График показывает нам группы людей в зависимости  от возраста, позволяет настроить таргет 
в зависимости от типа недвижимости под возрастную группу, наиболее интеерсующуюся этим типом недвижимости.

```python
Потенциально перспективные клиенты мужчиы 40-50 лет. Посмотрим для них цели покупки, медианный чек, тип недвжимости, площадь, 
```

```python
# Выделяем мужчин 40 - 50 лет
df_40_50 = df[
    (df["Age at time of purchase"] >= 40) &
    (df["Age at time of purchase"] <= 50)
].copy()

print("Number of clients aged 40–50:", df_40_50.shape[0])
print("Share of total sample: {:.1f}%".format(
    100 * df_40_50.shape[0] / df.shape[0]
))
```

```python
#6 - Purpose for men 40-50y.o.
plt.figure(figsize=(7, 4))
sns.countplot(
    x="Purpose",
    data=df_40_50,
    order=df_40_50["Purpose"].value_counts().index
)
plt.title("Purpose of purchase (age 40–50)")
plt.xlabel("Purpose")
plt.ylabel("Count")
plt.show()
```

```python
#7 - Median and mean price for men 40-50y.o.
price_stats_40_50 = df_40_50["Price_num"].agg(
    median_price="median",
    mean_price="mean"
)

price_stats_40_50

plt.figure(figsize=(7, 5))
sns.boxplot(y="Price_num", data=df_40_50)
plt.title("Price distribution (age 40–50)")
plt.ylabel("Price ($)")
plt.show()
```

```python
#8 - Real Estate Type

plt.figure(figsize=(7, 4))
sns.countplot(
    x="Type of property",
    data=df_40_50,
    order=df_40_50["Type of property"].value_counts().index
)
plt.title("Property type (age 40–50)")
plt.xlabel("Type of property")
plt.ylabel("Count")
plt.xticks(rotation=25)
plt.show()
```

```python
#9 - Sq M for men 40-50y.o.

area_stats_40_50 = (
    df_40_50
    .groupby("Purpose")["Area_sqm"]
    .agg(
        mean="mean",
        median="median",
        q25=lambda x: x.quantile(0.25),
        q75=lambda x: x.quantile(0.75),
        count="count"
    )
)

area_stats_40_50
```
